```python


learning_rate = 0.1
weight_update_ratio = 10
def sigmoid(x):
    return 1 / (1 + np.exp(-1*x))

def sigmoidderiv(x):
    return sigmoid(x) * (1 - sigmoid(x))

def upward_projection(pe, w):
    return sigmoidderiv(np.dot(w.T, pe)) 
def update_weight(predderiv, pe, u):
    return learning_rate * np.dot(pe * predderiv,  u.T)
    
def update_cause_unit(bu , pe):
    return learning_rate * np.subtract(bu, pe)

def top_down_prediction(w, td):
    return sigmoid(np.dot(w, td))

def prediction_error(u, pred):
    return np.subtract(u, pred)

def prediction_derivative(w,u):
    return sigmoidderiv(np.dot(w, u))
    
    



class Model():
  
    
    def __init__(self, layers=None):
        self.layers = layers or []
        self.upwards = []
        self.downwards = []
        
    def add(self, layer):
        self.layers.append(layer)
        
    def compile(self):
        for i in range(len(self.layers) + 1):
            self.upwards.append(None)
            self.dowards.append(None)
        
    def run(percept, top_down, epochs):
        percept = self.upwards[0]
        top_down = self.downwards[-1]
        for i,layer in enumerate(layers):
            #self.upwards[i] = layer.get_latents()
            self.downwards[i] = layer.get_latents() 
        for epoch in epochs:
            for i, layer in enumerate(self.layers):
                up, down = layer.run(self.upwards[i], downwards[i+1])
                self.upwards[i +1] = up
                self.downards[i] = down
class Layer():
    def __init__(self, latents, weights, weight_update_ratio, input_layer=False):
        self.latents = latents
        self.weights = weights
        self.weight_update_ratio = weight_update_ratio
        self.input_layer = input_layer
        
    def run(self, bottom_up, top_down,i):
        self.bottom_up = bottom_up
        self.top_down = top_down
        if self.input_layer:
            self.latents = self.bottom_up
            
        pred1 = top_down_prediction(self.weights, self.top_down)
        predderiv1 = prediction_derivative(self.weights, self.top_down) 
        up1 = upward_projection(pe1, self.weights)
        if not self.input_layer:
            self.latents += update_cause_unit(self.bottom_up, pe1)
        if i % self.weight_update_ratio == 0:
            self.weights += update_weight(predderiv1, pe1, self.top_down)
            
        return up1, self.latents 
        
        
    def get_latents(self):
        return self.latents
        
        
        


def two_layers():
    
    percept = np.array([0.1, 0.9])
    
    w1 = np.array([[0.1,0.3], [0.2,0.1]])
    w2 = np.array([[0.05, 0.05],[0.1,0.1]])
    u2 = np.array([0.1,0.1])
    u3 = np.array([0.5,0.5])

    pes = []
    pe2s = []
    weights = []
    preds = []

    percept = np.reshape(percept, (2,1))
    u2 = np.reshape(u2, (2,1))
    u3 = np.reshape(u3, (2,1))
    print(percept.shape)
    print(w1.shape)
    print(u2.shape)
    

    pe1s = []
    pe2s = []
    pe3s = []
    w1s = []
    w2s = []
    pred1s = []
    pred2s = []
    up1s = []
    up2s = []
    
    epochs = 5000
    

    for i in range(epochs):
        u1 = percept
        pred1 = top_down_prediction(w1, u2)
        predderiv1 = prediction_derivative(w1 , u2) 
        pe1 = prediction_error(u1, pred1)
        up1 = upward_projection(pe1, w1)
        pred2 = top_down_prediction(w2, u3)
        predderiv2 = prediction_derivative(w2, u3)
        pe2 = prediction_error(u2, pred2)
        up2 = upward_projection(pe2, w2)
        pe3 = prediction_error(u3, 0)
        #w1 += update_weight(predderiv1,pe1, u2)
        u2 += update_cause_unit(up1, pe2)
        #w2 += update_weight(predderiv2, pe2, u3)
        u3 += update_cause_unit(up2, pe3) 
        if i % weight_update_ratio == 0:
            w1 += update_weight(predderiv1,pe1, u2)
            w2 += update_weight(predderiv2, pe2, u3)


        #print("Pe: ", pe1)
        #print("Pe2: ", pe2)
        #print("Weight: ", w1)
        pe1s.append(np.sum(pe1))
        pe2s.append(np.sum(pe2))
        pe3s.append(np.sum(pe3))
        w1s.append(np.sum(w1))
        w2s.append(np.sum(w2))
        pred1s.append(np.sum(pred1))
        pred2s.append(np.sum(pred2))
        up1s.append(np.sum(up1))
        up2s.append(np.sum(up2))
     
    
    for i in range(epochs):
        u1 = np.array([0.9, 0.1])
        u1 = np.reshape(u1, (2,1))
        pred1 = top_down_prediction(w1, u2)
        predderiv1 = prediction_derivative(w1 , u2) 
        pe1 = prediction_error(u1, pred1)
        up1 = upward_projection(pe1, w1)
        pred2 = top_down_prediction(w2, u3)
        predderiv2 = prediction_derivative(w2, u3)
        pe2 = prediction_error(u2, pred2)
        up2 = upward_projection(pe2, w2)
        pe3 = prediction_error(u3, 0)
        #w1 += update_weight(predderiv1,pe1, u2)
        u2 += update_cause_unit(up1, pe2)
        #w2 += update_weight(predderiv2, pe2, u3)
        u3 += update_cause_unit(up2, pe3) 
        if i % weight_update_ratio == 0:
            w1 += update_weight(predderiv1,pe1, u2)
            w2 += update_weight(predderiv2, pe2, u3)


        #print("Pe: ", pe1)
        #print("Pe2: ", pe2)
        #print("Weight: ", w1)
        pe1s.append(np.sum(pe1))
        pe2s.append(np.sum(pe2))
        pe3s.append(np.sum(pe3))
        w1s.append(np.sum(w1))
        w2s.append(np.sum(w2))
        pred1s.append(np.sum(pred1))
        pred2s.append(np.sum(pred2))
        up1s.append(np.sum(up1))
        up2s.append(np.sum(up2))
        
    for i in range(epochs):
        u1 = percept
        pred1 = top_down_prediction(w1, u2)
        predderiv1 = prediction_derivative(w1 , u2) 
        pe1 = prediction_error(u1, pred1)
        up1 = upward_projection(pe1, w1)
        pred2 = top_down_prediction(w2, u3)
        predderiv2 = prediction_derivative(w2, u3)
        pe2 = prediction_error(u2, pred2)
        up2 = upward_projection(pe2, w2)
        pe3 = prediction_error(u3, 0)
        #w1 += update_weight(predderiv1,pe1, u2)
        u2 += update_cause_unit(up1, pe2)
        #w2 += update_weight(predderiv2, pe2, u3)
        u3 += update_cause_unit(up2, pe3) 
        if i % weight_update_ratio == 0:
            w1 += update_weight(predderiv1,pe1, u2)
            w2 += update_weight(predderiv2, pe2, u3)


        #print("Pe: ", pe1)
        #print("Pe2: ", pe2)
        #print("Weight: ", w1)
        pe1s.append(np.sum(pe1))
        pe2s.append(np.sum(pe2))
        pe3s.append(np.sum(pe3))
        w1s.append(np.sum(w1))
        w2s.append(np.sum(w2))
        pred1s.append(np.sum(pred1))
        pred2s.append(np.sum(pred2))
        up1s.append(np.sum(up1))
        up2s.append(np.sum(up2))
        
    for i in range(epochs):
        u1 = np.array([0.3,0.7])
        u1 = np.reshape(u1, (2,1))
        predderiv1 = prediction_derivative(w1 , u2) 
        pe1 = prediction_error(u1, pred1)
        up1 = upward_projection(pe1, w1)
        pred2 = top_down_prediction(w2, u3)
        predderiv2 = prediction_derivative(w2, u3)
        pe2 = prediction_error(u2, pred2)
        up2 = upward_projection(pe2, w2)
        pe3 = prediction_error(u3, 0)
        #w1 += update_weight(predderiv1,pe1, u2)
        u2 += update_cause_unit(up1, pe2)
        #w2 += update_weight(predderiv2, pe2, u3)
        u3 += update_cause_unit(up2, pe3) 
        if i % weight_update_ratio == 0:
            w1 += update_weight(predderiv1,pe1, u2)
            w2 += update_weight(predderiv2, pe2, u3)
        
        pe1s.append(np.sum(pe1))
        pe2s.append(np.sum(pe2))
        pe3s.append(np.sum(pe3))
        w1s.append(np.sum(w1))
        w2s.append(np.sum(w2))
        pred1s.append(np.sum(pred1))
        pred2s.append(np.sum(pred2))
        up1s.append(np.sum(up1))
        up2s.append(np.sum(up2))
            
    for i in range(epochs):
        u1 = np.array([0.5, 0.5])
        u1 = np.reshape(u1, (2,1))
        predderiv1 = prediction_derivative(w1 , u2) 
        pe1 = prediction_error(u1, pred1)
        up1 = upward_projection(pe1, w1)
        pred2 = top_down_prediction(w2, u3)
        predderiv2 = prediction_derivative(w2, u3)
        pe2 = prediction_error(u2, pred2)
        up2 = upward_projection(pe2, w2)
        pe3 = prediction_error(u3, 0)
        #w1 += update_weight(predderiv1,pe1, u2)
        u2 += update_cause_unit(up1, pe2)
        #w2 += update_weight(predderiv2, pe2, u3)
        u3 += update_cause_unit(up2, pe3) 
        if i % weight_update_ratio == 0:
            w1 += update_weight(predderiv1,pe1, u2)
            w2 += update_weight(predderiv2, pe2, u3)

        pe1s.append(np.sum(pe1))
        pe2s.append(np.sum(pe2))
        pe3s.append(np.sum(pe3))
        w1s.append(np.sum(w1))
        w2s.append(np.sum(w2))
        pred1s.append(np.sum(pred1))
        pred2s.append(np.sum(pred2))
        up1s.append(np.sum(up1))
        up2s.append(np.sum(up2))
            
    l0 = np.copy(u3)
                
        
    
    print("Beginning latents!")
    for i in range(1000):
        u1 = np.array([0.3,0.7])
        u1 = np.reshape(u1, (2,1))
        predderiv1 = prediction_derivative(w1 , u2)
        pe1 = prediction_error(u1, pred1)
        up1 = upward_projection(pe1, w1)
        pred2 = top_down_prediction(w2, u3)
        predderiv2 = prediction_derivative(w2, u3)
        pe2 = prediction_error(u2, pred2)
        up2 = upward_projection(pe2, w2)
        pe3 = prediction_error(u3, 0)
        #w1 += update_weight(predderiv1,pe1, u2)
        u2 += update_cause_unit(up1, pe2)
        #w2 += update_weight(predderiv2, pe2, u3)
        u3 += update_cause_unit(up2, pe3) 
        #print(u3)
        l1 = np.copy(u3)
        
    for i in range(1000):
        u1 = np.array([0.5, 0.5])
        u1 = np.reshape(u1, (2,1))
        predderiv1 = prediction_derivative(w1 , u2)
        pe1 = prediction_error(u1, pred1)
        up1 = upward_projection(pe1, w1)
        pred2 = top_down_prediction(w2, u3)
        predderiv2 = prediction_derivative(w2, u3)
        pe2 = prediction_error(u2, pred2)
        up2 = upward_projection(pe2, w2)
        pe3 = prediction_error(u3, 0)
        #w1 += update_weight(predderiv1,pe1, u2)
        u2 += update_cause_unit(up1, pe2)
        #w2 += update_weight(predderiv2, pe2, u3)
        u3 += update_cause_unit(up2, pe3)
        #print(u3)
        l2 = np.copy(u3)

    for i in range(1000):
        u1 = np.array([0.9, 0.1])
        u1 = np.reshape(u1, (2,1))
        predderiv1 = prediction_derivative(w1 , u2) 
        pe1 = prediction_error(u1, pred1)
        up1 = upward_projection(pe1, w1)
        pred2 = top_down_prediction(w2, u3)
        predderiv2 = prediction_derivative(w2, u3)
        pe2 = prediction_error(u2, pred2)
        up2 = upward_projection(pe2, w2)
        pe3 = prediction_error(u3, 0)
        #w1 += update_weight(predderiv1,pe1, u2)
        u2 += update_cause_unit(up1, pe2)
        #w2 += update_weight(predderiv2, pe2, u3)
        u3 += update_cause_unit(up2, pe3) 
        #print(u3)
        l3 = np.copy(u3)

    for i in range(1000):
        u1 = np.array([0.1, 0.9])
        u1 = np.reshape(u1, (2,1))
        predderiv1 = prediction_derivative(w1 , u2) 
        pe1 = prediction_error(u1, pred1)
        up1 = upward_projection(pe1, w1)
        pred2 = top_down_prediction(w2, u3)
        predderiv2 = prediction_derivative(w2, u3)
        pe2 = prediction_error(u2, pred2)
        up2 = upward_projection(pe2, w2)
        pe3 = prediction_error(u3, 0)
        #w1 += update_weight(predderiv1,pe1, u2)
        u2 += update_cause_unit(up1, pe2)
        #w2 += update_weight(predderiv2, pe2, u3)
        u3 += update_cause_unit(up2, pe3) 
       # print(u3)
        l4 = np.copy(u3)
        

        #print("Pe: ", pe1)
        #print("Pe2: ", pe2)
        #print("Weight: ", w1)
        pe1s.append(np.sum(pe1))
        pe2s.append(np.sum(pe2))
        pe3s.append(np.sum(pe3))
        w1s.append(np.sum(w1))
        w2s.append(np.sum(w2))
        pred1s.append(np.sum(pred1))
        pred2s.append(np.sum(pred2))
        up1s.append(np.sum(up1))
        up2s.append(np.sum(up2))
        
    for i in range(1000):
        u0 = np.array([0,0])
        u0 = np.reshape(u0, (2,1))
        u3 = l4
       # prop2_3 = top_down_prediction(w2, u3)
        #prop1_3 = top_down_prediction(w1, prop2_3)
        pred1 = top_down_prediction(w1, u2)
        predderiv1 = prediction_derivative(w1 , u2)
        pe1 = prediction_error(u1, pred1)
        up1 = upward_projection(pe1, w1)
        pred2 = top_down_prediction(w2, u3)
        predderiv2 = prediction_derivative(w2, u3)
        pe2 = prediction_error(u2, pred2)
        up2 = upward_projection(pe2, w2)
        pe3 = prediction_error(u3, 0)
        #w1 += update_weight(predderiv1,pe1, u2)
        u2 += update_cause_unit(up1, pe2)
        u1 += update_cause_unit(u0, pe1)
    prop1_4 = np.copy(u1)

    for i in range(1000):
        u0 = np.array([0,0])
        u0 = np.reshape(u0, (2,1))
        u3 = l3
       # prop2_3 = top_down_prediction(w2, u3)
        #prop1_3 = top_down_prediction(w1, prop2_3)
        pred1 = top_down_prediction(w1, u2)
        predderiv1 = prediction_derivative(w1 , u2)
        pe1 = prediction_error(u1, pred1)
        up1 = upward_projection(pe1, w1)
        pred2 = top_down_prediction(w2, u3)
        predderiv2 = prediction_derivative(w2, u3)
        pe2 = prediction_error(u2, pred2)
        up2 = upward_projection(pe2, w2)
        pe3 = prediction_error(u3, 0)
        #w1 += update_weight(predderiv1,pe1, u2)
        u2 += update_cause_unit(up1, pe2)
        u1 += update_cause_unit(u0, pe1)
    prop1_3 = np.copy(u1)

    for i in range(1000):
        u0 = np.array([0,0])
        u0 = np.reshape(u0, (2,1))
        u3 = l2
       # prop2_3 = top_down_prediction(w2, u3)
        #prop1_3 = top_down_prediction(w1, prop2_3)
        pred1 = top_down_prediction(w1, u2)
        predderiv1 = prediction_derivative(w1 , u2) 
        pe1 = prediction_error(u1, pred1)
        up1 = upward_projection(pe1, w1)
        pred2 = top_down_prediction(w2, u3)
        predderiv2 = prediction_derivative(w2, u3)
        pe2 = prediction_error(u2, pred2)
        up2 = upward_projection(pe2, w2)
        pe3 = prediction_error(u3, 0)
        #w1 += update_weight(predderiv1,pe1, u2)
        u2 += update_cause_unit(up1, pe2)
        p1 = top_down_prediction(w1, u2)
    prop1_2 = np.copy(p1)

    for i in range(1000):
        u0 = np.array([0,0])
        u0 = np.reshape(u0, (2,1))
        u3 = l1
       # prop2_3 = top_down_prediction(w2, u3)
        #prop1_3 = top_down_prediction(w1, prop2_3)
        pred1 = top_down_prediction(w1, u2)
        predderiv1 = prediction_derivative(w1 , u2) 
        pe1 = prediction_error(u1, pred1)
        up1 = upward_projection(pe1, w1)
        pred2 = top_down_prediction(w2, u3)
        predderiv2 = prediction_derivative(w2, u3)
        pe2 = prediction_error(u2, pred2)
        up2 = upward_projection(pe2, w2)
        pe3 = prediction_error(u3, 0)
        #w1 += update_weight(predderiv1,pe1, u2)
        u2 += update_cause_unit(up1, pe2)
        p1 = top_down_prediction(w1, u2)
    prop1_1 = np.copy(p1)

    for i in range(1000):
        u0 = np.array([0,0])
        u0 = np.reshape(u0, (2,1))
        u3 = l4
        #print(u3)
       # prop2_3 = top_down_prediction(w2, u3)
        #prop1_3 = top_down_prediction(w1, prop2_3)
        #pred1 = top_down_prediction(w1, u2)
        #predderiv1 = prediction_derivative(w1 , u2) 
        #pe1 = prediction_error(u1, pred1)
        #up1 = upward_projection(pe1, w1)
        pred2 = top_down_prediction(w2, u3)
       # predderiv2 = prediction_derivative(w2, u3)
        pe2 = prediction_error(u2, pred2)
        #up2 = upward_projection(pe2, w2)
        #pe3 = prediction_error(u3, 0)
        #w1 += update_weight(predderiv1,pe1, u2)
        delta =  update_cause_unit(up1, pe2)
        u2 += delta
        print(up1)
        print(pe2)
        print(np.subtract(up2, pe2))
        print(update_cause_unit(up1, pe2))
        p1 = top_down_prediction(w1, u2)
    prop1_0 = np.copy(p1)
        
        


    pe1s = np.array(pe1s)
    pe2s = np.array(pe2s)
    pe3s = np.array(pe3s)
    w1s = np.array(w1s)
    w2s = np.array(w2s)
    pred1s = np.array(pred1s)
    pred2s = np.array(pred2s)
    up1s = np.array(up1s)
    up2s = np.array(up2s)
    #print("\n")
    #print(pe1s)
    #print(pe1s.shape)
    #print(pe1s[0])
    #print(type(pe1s[0]))
    plt.plot(pe1s)
    plt.title("PE 1")
    plt.show()
    plt.plot(pe2s)
    plt.title("PE2")
    plt.show()
    plt.plot(pe3s)
    plt.title("PE3")
    plt.show()
    plt.plot(w1s)
    plt.title("W1")
    plt.show()
    plt.plot(w2s)
    plt.title("W2")
    plt.show()
    plt.plot(pred1s)
    plt.title("Pred1")
    plt.show()
    plt.plot(pred2s)
    plt.title("Pred2")
    plt.show()
    plt.plot(up1s)
    plt.title("Up1")
    plt.show()
    plt.plot(up2s)
    plt.title("Up2")
    plt.show()
    
    print("\n")
    print("Latents:")
    print(l0)
    print(l1)
    print(l2)
    print(l3)
    print(l4)
    print("Predictions")
    print(prop1_0)
    print(prop1_1)
    print(prop1_2)
    print(prop1_3)
    print(prop1_4)
    
    
    
two_layers()
```
